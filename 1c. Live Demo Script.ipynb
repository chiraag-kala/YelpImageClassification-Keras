```python
#Importing Libraries
import numpy as np
import os
import cv2
import os
import os.path
import matplotlib.pyplot as plt
%matplotlib inline
import keras
import tensorflow
from keras.models import load_model
```

```python
pics_directory = '/Users/Monu/Desktop/Deep Learning and Image Recognition/Deep Learning Final Project/pics2/'
model_directory = '/Users/Monu/Desktop/Deep Learning and Image Recognition/Deep Learning Final Project/'
model_name = 'final.h5'
```

```python
images = []

for dirpath, dirnames, filenames in os.walk(pics_directory):
    for filename in filenames:
        filepath = pics_directory + filename
        image = cv2.imread(filepath)
        image = cv2.resize(cv2.cvtColor(image, cv2.COLOR_BGR2RGB),(256,256))
        images.append(image)
        
img_array = np.vstack(images)
img_array = img_array.reshape(-1, 256, 256, 3)
img_array = img_array.transpose(0, 1, 2, 3)
```
