### Importing Libraries

```python
#Importing Libraries
import numpy as np
import os
import cv2
import os
import os.path
import matplotlib.pyplot as plt
%matplotlib inline
```

### Defining the Directories

```python
drink_directory = '/Users/Monu/yelp_data/drink/'
food_directory = '/Users/Monu/yelp_data/food/'
inside_directory = '/Users/Monu/yelp_data/inside/'
outside_directory = '/Users/Monu/yelp_data/outside/'
```

### 1. Creating and Pickling Drinks Array

```python
drink_images = []

#number of total images to load
n = 11000 

#batch size
batch = 1000 

image_index = range(0,n,batch)

for i in image_index:
    for dirpath, dirnames, filenames in os.walk(drink_directory):
        for filename in [f for f in filenames][i:i+1000]:
            filepath = drink_directory + "/" + filename
            image = cv2.imread(filepath)
            image = cv2.resize(cv2.cvtColor(image, cv2.COLOR_BGR2RGB),(256,256))
            drink_images.append(image)
    print(i)

drink_img_array = np.vstack(drink_images)
drink_img_array = drink_img_array.reshape(-1, 256, 256, 3)
drink_img_array = drink_img_array.transpose(0, 1, 2, 3)
drink_img_array = drink_img_array/255
```

```python
np.save('drink_reduced', drink_img_array)
```
